### O problema dos prisioneiros contra um bot aleatório (1)

O usuário irá escolher uma estratégia (trair ou cooperar) no dilema dos prisioneiros. Uma estratégia aleatória será escolhida pelo programa. Aparecerá para o usuário qual estratégia ele escolheu, qual foi sorteada pelo programa e qual o resultado daquele jogo. Além disso, aparecerá um "placar geral", em que aparecerá as estratégias adotadas pelo usuário até o momento e o "score" dele.

```python
def randomico():
    T = 0
    C = 0
    t = 0 
    n = 0
    score = 0 
    def uma_vez():
        change = 0 
        randomico = ["trair", "cooperar"]
        user_choice = (input("Você irá trair ou cooperar?")).lower()
        machine_choice = random.choice(randomico)
        if user_choice == "trair":
            if machine_choice == "trair": #TRAIR x TRAIR
                change -= 5
            else: #TRAIR x COOPERAR
                change -= 1
        elif user_choice == "cooperar": 
            if machine_choice == "trair": #COOPERAR x TRAIR
                change -= 10
            else: #COOPERAR x COOPERAR 
                change -= 2
        else:
            return "X"
        result = f'Você escolheu {user_choice} e o outro prisioneiro escolheu {machine_choice}'
        print(result)
        return [change, user_choice]
    while n < 9:
        if t < 5:
            for i in uma_vez():
                if i == "X": return "Entrada Inválida"
                if i == -5 or i == -1 or i == -10 or i == -2: score += i
                if i == "trair": 
                    T += 1 
                    t += 1
                if i == "cooperar": 
                    C += 1  
                    t -= t
            n += 1
        print(f'Seu score atual é {score}')
        print(f'Trair: {T}  |  Cooperar: {C}')
        print("-------------------------------------------------------------------------------------")
        if t == 5: 
             return f'PARABÉNS!! Você percebeu que a estratégia ótima para esse problema é sempre trair.'
    return f'Vish, você não conseguiu percebecer a estratégia ótima'
```

### Ocorrências repetidas - objetivo BÔNUS

#### Contra diferentes "tipos de jogadores", escolhido pelo usuário

O comportamento do usuário varia conforme o jogador que ele está enfrentando - o usuário escolhe contra quem ele vai jogar e adota um comportamento ao escolher se coopera ou trai

#### Contra diferentes "tipos de jogadores", mas "no escuro"

O comportamento do usuário varia conforme o jogador que ele está enfrentando, mas o usuário NÃO escolhe contra quem ele vai jogar, então deve ter um tipo de memória

```python
import pygame

# Inicialização
pygame.init()

# Cores
BRANCO = (255, 255, 255)
PRETO = (0, 0, 0)
CINZA = (200, 200, 200)
AZUL = (50, 150, 250)

# Configurações da Tela
LARGURA, ALTURA = 600, 500
tela = pygame.display.set_mode((LARGURA, ALTURA))
pygame.display.set_caption("Exemplo de Tabela no Pygame")

# Configurações da Tabela
LINHAS = 3
COLUNAS = 3
LARGURA_CELULA = 150
ALTURA_CELULA = 50
MARGEM_X = (LARGURA - (COLUNAS * LARGURA_CELULA)) // 2
MARGEM_Y = (ALTURA - (LINHAS * ALTURA_CELULA)) // 2

# Matriz de dados (exemplo)
dados = [
    ["ID", "Nome", "Status"],
    ["1", "Pygame", "Ativo"],
    ["2", "Python", "Pronto"]
]

fonte = pygame.font.SysFont(None, 24)

# Loop principal
rodando = True
while rodando:
    for evento in pygame.event.get():
        if evento.type == pygame.QUIT:
            rodando = False

    tela.fill(BRANCO)

    # Desenhar a tabela
    for i in range(LINHAS):
        for j in range(COLUNAS):
            # Coordenadas da célula
            x = MARGEM_X + j * LARGURA_CELULA
            y = MARGEM_Y + i * ALTURA_CELULA
            
            # Criar retângulo e desenhar borda
            retangulo = pygame.Rect(x, y, LARGURA_CELULA, ALTURA_CELULA)
            pygame.draw.rect(tela, CINZA, retangulo, 1)  # Borda
            
            # Preencher fundo da primeira linha para destacar o cabeçalho
            if i == 0:
                pygame.draw.rect(tela, AZUL, retangulo)
            
            # Renderizar o texto
            texto = fonte.render(dados[i][j], True, PRETO if i > 0 else BRANCO)
            
            # Centralizar o texto na célula
            texto_rect = texto.get_rect(center=retangulo.center)
            tela.blit(texto, texto_rect)

    pygame.display.flip()

pygame.quit()
```
