## CONFECÇÃO:

### O problema dos prisioneiros contra um bot aleatório (1)

O usuário irá escolher uma estratégia (trair ou cooperar) no dilema dos prisioneiros. Uma estratégia aleatória será escolhida pelo programa. Aparecerá para o usuário qual estratégia ele escolheu, qual foi sorteada pelo programa e qual o resultado daquele jogo. Além disso, aparecerá um "placar geral", em que aparecerá as estratégias adotadas pelo usuário até o momento e o "score" dele.

```python
def randomico():
    T = 0
    C = 0
    t = 0 
    n = 0
    score = 0 
    def uma_vez():
        change = 0 
        randomico = ["trair", "cooperar"]
        user_choice = (input("Você irá trair ou cooperar?")).lower()
        machine_choice = random.choice(randomico)
        if user_choice == "trair":
            if machine_choice == "trair": #TRAIR x TRAIR
                change -= 5
            else: #TRAIR x COOPERAR
                change -= 1
        elif user_choice == "cooperar": 
            if machine_choice == "trair": #COOPERAR x TRAIR
                change -= 10
            else: #COOPERAR x COOPERAR 
                change -= 2
        else:
            return "X"
        result = f'Você escolheu {user_choice} e o outro prisioneiro escolheu {machine_choice}'
        print(result)
        return [change, user_choice]
    while n < 9:
        if t < 5:
            for i in uma_vez():
                if i == "X": return "Entrada Inválida"
                if i == -5 or i == -1 or i == -10 or i == -2: score += i
                if i == "trair": 
                    T += 1 
                    t += 1
                if i == "cooperar": 
                    C += 1  
                    t -= t
            n += 1
        print(f'Seu score atual é {score}')
        print(f'Trair: {T}  |  Cooperar: {C}')
        print("-------------------------------------------------------------------------------------")
        if t == 5: 
             return f'PARABÉNS!! Você percebeu que a estratégia ótima para esse problema é sempre trair.'
    return f'Vish, você não conseguiu percebecer a estratégia ótima'
```
